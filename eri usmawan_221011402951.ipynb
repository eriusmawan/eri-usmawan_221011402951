```python
# List of company details
company_detail_list = [
    {
        'name': 'Company 1',
        'domain': 'Retail',
        'country': 'United States'
    },
    {
        'name': 'Company 2',
        'domain': 'Technology',
        'country': 'United Kingdom'
    },
    {
        'name': 'Company 3',
        'domain': 'Healthcare',
        'country': 'United States'
    }
]

# List of employee details
employee_detail_list = [
    {
        'name': 'EMP-0001',
        'first_name': 'John',
        'last_name': 'Doe',
        'full_name': 'John Doe',
        'company': 'Company 1',
        'nationality': 'Australia'
    },
    {
        'name': 'EMP-0002',
        'first_name': 'Tom',
        'last_name': 'Smith',
        'full_name': 'Tom Smith',
        'company': 'Company 2',
        'nationality': 'United States'
    },
    {
        'name': 'EMP-0003',
        'first_name': 'Andrew',
        'last_name': 'Sebastian',
        'full_name': 'Andrew Sebastian',
        'company': 'Company 3',
        'nationality': 'United States'
    },
    {
        'name': 'EMP-0005',
        'first_name': 'Ying Han',
        'last_name': 'Tan',
        'full_name': 'Ying Han Tan',
        'company': 'Company 1',
        'nationality': 'Australia'
    },
    {
        'name': 'EMP-0015',
        'first_name': 'Kenneth',
        'last_name': 'Ng',
        'full_name': 'Kenneth Ng',
        'company': 'Company 3',
        'nationality': 'United States'
    },
    {
        'name': 'EMP-0018',
        'first_name': 'Rubby',
        'last_name': 'Lee',
        'full_name': 'Rubby Lee',
        'company': 'Company 2',
        'nationality': 'Hong Kong'
    },
    {
        'name': 'EMP-0017',
        'first_name': 'Robert',
        'last_name': 'White',
        'full_name': 'Robert White',
        'company': 'Company 1',
        'nationality': 'United Kingdom'
    }
]

def get_company_employee_nationalities(companies, employees):
    # Create a dictionary to hold the company employee nationality counts
    company_nationality_counts = {}

    for company in companies:
        company_name = company['name']
        company_nationality_counts[company_name] = {}

    for employee in employees:
        company_name = employee['company']
        nationality = employee['nationality']
        if nationality in company_nationality_counts[company_name]:
            company_nationality_counts[company_name][nationality] += 1
        else:
            company_nationality_counts[company_name][nationality] = 1

    # Create the result list with the required information
    result = []
    for company in companies:
        company_name = company['name']
        result.append({
            'company': company_name,
            'employee_nationality': company_nationality_counts[company_name]
        })

    return result

# Get the companies with the number of employee nationalities
result = get_company_employee_nationalities(company_detail_list, employee_detail_list)

# Print the result
print(result)
```
